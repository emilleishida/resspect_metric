```python
# Color map
rainbow = cm = plt.get_cmap('plasma_r')
cNorm  = colors.LogNorm(vmin=1, vmax=52) #colors.Normalize(vmin=0, vmax=50)
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=rainbow)
color_map = scalarMap.to_rgba(np.arange(1, 52))
```

```python
fname = '/media2/RESSPECT2/data/posteriors_ddf/omprior_0.01_flat/summary_cases_emille.csv' 

df = pd.read_csv(fname)
```

```python
change_names = {}
change_names['SN-II'] = 'SNII'
change_names['SN-Ibc'] = 'SNIbc'
change_names['SN-Iax'] = 'SNIax'
change_names['SN-Ia-91bg'] = 'SNIa-91bg'
change_names['SNLS-I'] = 'SLSN-I'

pairs = [['fom3', 'efficiency'], ['accuracy', 'fom3'], ['fom3', 'purity'],
         ['accuracy', 'fom3'], ['accuracy', 'purity'], ['efficiency', 'purity']]


ax = {}
p = []

plt.figure(figsize=(24, 10))

for i in range(6):

    ax[i] = plt.subplot(2,3, i + 1)

    flag_perfect = df['case'].values == 'perfect3000'
    plt.scatter(df[pairs[i][0]][flag_perfect], df[pairs[i][1]][flag_perfect], color='k', marker='*', s=100, label=a)

    flag_random = df['case'].values == 'random3000'
    plt.scatter(df[pairs[i][0]][flag_random], df[pairs[i][1]][flag_random], color='tab:red', marker='*', s=100, label=a)

    flag_fiducial = df['case'].values == 'fiducial3000'
    plt.scatter(df[pairs[i][0]][flag_fiducial], df[pairs[i][1]][flag_fiducial], color='tab:blue', marker='*', s=100, label=a)

    for j, (a, c) in enumerate(zip(remap_dict, color_nums)):
        if a not in ['perfect3000', 'random3000', 'fiducial3000'] and a in df['case'].values:
        
            class_ = str.split(remap_dict[a])[0]        
            if class_ in change_names.keys():
                class_ = change_names[class_]
            
            mfc = color_map[c]
            flag_case = df['case'].values == a
        
            if i == 5:
                p1, = plt.plot(df[flag_case][pairs[i][0]], df.iloc[flag_case][pairs[i][1]], color=color_map[c], ms=10, marker=all_shapes[class_], mfc=mfc,
                        label=a)
                p.append(p1)
            else:
                plt.plot(df[flag_case][pairs[i][0]], df.iloc[flag_case][pairs[i][1]], color=color_map[c], ms=10, marker=all_shapes[class_], mfc=mfc,
                        label=a)
        
    plt.xlabel(pairs[i][0], fontsize=13)
    plt.ylabel(pairs[i][1], fontsize=13)
    
    if i == 5:
        fontP = FontProperties()
        fontP.set_size(15)
        plt.legend(handles=p, bbox_to_anchor=(1.05, 2.2), loc='upper left', prop=fontP, frameon=False,  labelspacing=1)

plt.savefig('classification_metrics.png',  bbox_inches = 'tight', pad_inches = 0.25)
```
