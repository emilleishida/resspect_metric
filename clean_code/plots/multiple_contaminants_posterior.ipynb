```python
import matplotlib.pylab as plt
import pandas as pd
import numpy as np

from scipy import stats
import matplotlib.colors as colors
import matplotlib.cm as cmx
```

```python
data_dir = '/media/RESSPECT/data/PLAsTiCC/for_metrics/final_data/'


posteriors = {}
density = {}
for field in ['DDF', 'WFD']:
    
    density[field] = {}
    posteriors[field] = {}
    
    for size in [1500, 3000, 6000]:
        
        posteriors[field][size] = {}
        posteriors[field][size]['fiducial'] = pd.read_csv(data_dir + field + \
                                                '/v2/posteriors/csv/chains_fiducial' + str(size) + '_lowz_withbias.csv')
        posteriors[field][size]['random'] = pd.read_csv(data_dir + field + \
                                                '/v2/posteriors/csv/chains_random' + str(size) + '_lowz_withbias.csv')
        posteriors[field][size]['perfect'] = pd.read_csv(data_dir + field + \
                                                '/v2/posteriors/csv/chains_perfect' + str(size) + '_lowz_withbias.csv')
    
        density[field][size] = {}
        density[field][size]['fiducial'] = stats.kde.gaussian_kde(posteriors[field][size]['fiducial']['w'], bw_method=0.3)
        density[field][size]['random'] = stats.kde.gaussian_kde(posteriors[field][size]['random']['w'], bw_method=0.3)
        density[field][size]['perfect'] = stats.kde.gaussian_kde(posteriors[field][size]['perfect']['w'], bw_method=0.3)
```

```python
x = np.arange(-1.15, -0.95, .0001)

plt.figure(figsize=(10,10))
ax1 = plt.subplot(2,1,1)

plt.plot(x, density['DDF'][1500]['perfect'](x), color='black', ls=':')
plt.plot(x, density['DDF'][1500]['random'](x), color='#d62728', ls=':')
plt.plot(x, density['DDF'][1500]['fiducial'](x), color='#1f77b4', ls=':')

plt.plot(x, density['DDF'][3000]['perfect'](x), color='black')
plt.plot(x, density['DDF'][3000]['random'](x), color='#d62728')
plt.plot(x, density['DDF'][3000]['fiducial'](x), color='#1f77b4')

plt.plot(x, density['DDF'][6000]['perfect'](x), color='black', ls='--')
plt.plot(x, density['DDF'][6000]['random'](x), color='#d62728', ls='--')
plt.plot(x, density['DDF'][6000]['fiducial'](x), color='#1f77b4', ls='--')

l1 = ax1.plot([], [], color='gray', ls=':', label=1500, lw=3)
l2 = ax1.plot([], [], color='gray', label=3000, lw=3)
l3 = ax1.plot([], [], color='gray', ls='--', label=6000, lw=3)
l4 = ax1.plot([], [], color='white', label='   ', lw=3)
l5 = ax1.scatter([], [], color='black', marker='o', label='perfect', s=30)
l6 = ax1.scatter([], [], color='#d62728', marker='o', label='random', s=30)
l7 = ax1.scatter([], [], color='#1f77b4', marker='o', label='fiducial', s=30)

plt.xticks([])
plt.yticks([])
plt.text(-0.965, 51, 'DDF', fontsize=18)
plt.ylabel('PDF', fontsize=18)

handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles, frameon=False, bbox_to_anchor=(1.3, 0.32), labelspacing=0.5, fontsize=16)


plt.subplot(2,1,2)

plt.plot(x, density['WFD'][1500]['perfect'](x), color='black', ls=':')
plt.plot(x, density['WFD'][1500]['random'](x), color='#d62728', ls=':')
plt.plot(x, density['WFD'][1500]['fiducial'](x), color='#1f77b4', ls=':')

plt.plot(x, density['WFD'][3000]['perfect'](x), color='black')
plt.plot(x, density['WFD'][3000]['random'](x), color='#d62728')
plt.plot(x, density['WFD'][3000]['fiducial'](x), color='#1f77b4')

plt.plot(x, density['WFD'][6000]['perfect'](x), color='black', ls='--')
plt.plot(x, density['WFD'][6000]['random'](x), color='#d62728', ls='--')
plt.plot(x, density['WFD'][6000]['fiducial'](x), color='#1f77b4', ls='--')

plt.text(-0.965, 32, 'WFD', fontsize=18)
plt.ylabel('PDF', fontsize=18)
plt.xlabel('w', fontsize=18)
plt.yticks([])

plt.subplots_adjust(hspace=0.001, top=0.99, right=0.8)

plt.savefig('dists_null.png')
#plt.show()
```
